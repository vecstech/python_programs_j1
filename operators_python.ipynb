# operators in python

## python have 6 types of operators
1. arithematic operators
2. assignment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators


```python
#arithematic operators - +,-,*,/,//,%
a=10
b=20
c=30
d=40
print(a+b)
print(b-d)
print(a*c)
print(b/a)
print(b//a)
```

```python
#assigmnet operators - +=, -=,*=,/=,//=,%=
a=10
a=a+10
print(a)
```

```python
a=10
a+=10
print(a)
```

```python
a=25
a-=50
print(a)
```

```python
b=15
b*=25
print(b)
```

```python
c=100
c/=15
print(c)
```

```python
#comparision operators - >,<,>=,<=,==,!= (return True or false)
a=5
b=10
c=10
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
# logical operator - and, or, not
# and - this operators checking all conditions are true result is true rest cases failed
# or - this operator check either one condtions true true then resukt treu rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
print(not a>b or b!=c)
```

```python
# membership operator - checking members are present in that sequence or collection - in , not in
names=["shiva","vikas","divya","alok"]
print("alok" in names)
print("chandra" in names)
print("Shiva" not in names)
print("Vikas" in names)
```

```python
# identity operators -  memory identification of values - is, is not
a=10000
b=10000
print(id(a))
print(id(b))
print(a is b)
```

```python
a=10
b=10
print(id(a))
print(id(b))
print(a is b)
```
